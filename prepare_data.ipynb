```python
def get_md5(file_name):
    hash_md5 = hashlib.md5()
    with open(file_name, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    md5 = hash_md5.hexdigest()
    return md5
```

```python
# check duplicate file 
def get_list_md5(folder):
    list_md5 = []
    list_path =[]
    for file in os.listdir(folder):
        path = os.path.join(folder, file)
        md5 = get_md5(path)
        list_md5.append([md5])
        list_path.append([path])
    return [list_md5, list_path];
```

```python
fire_dir = "Data\\Fire images"
fire_md5_list,path_list_fire = get_list_md5(fire_dir)
```

```python
def check_dup(list_md5, path_list):
    list_length = len(list_md5)
    remove_list = []
    flag = 0
    file_remove = 0
    for i, md5_code in enumerate(list_md5):
        flag = 0
        for j in range(0, len(remove_list)):
            if md5_code == remove_list[j][1]:
                flag = 1
                
        if flag == 0:
            counter = 0
            for k in range(i+1, list_length):
                if md5_code == list_md5[k]:
                    counter += 1
                    file_remove +=1
                    os.remove(path_list[k][0])
            if counter > 0:
                remove_list.append([[counter], list_md5[i]])
                
    print("File Removed:", )
    return remove_list
```

```python
list_removed = check_dup(fire_md5_list, path_list_fire)
print(list_removed)
```

```python
list_removed = check_dup(normal_md5_list, path_list_nor)
print(list_removed)
```

```python
def img_to_npy(dir_1, dir_2, img_size):
    data = []
    for img in os.listdir(dir_1):
        try:
            path = os.path.join(dir_1, img)
            # print(path)
            img = cv2.imread(path, cv2.IMREAD_COLOR)
            img = cv2.resize(img, (img_size, img_size))
            data.append([np.array(img), np.array([1])])
        except:
            pass
    for img in os.listdir(dir_2):
        try:
            path = os.path.join(dir_2, img)
            img = cv2.imread(path, cv2.IMREAD_COLOR)
            img = cv2.resize(img, (img_size, img_size))
            data.append([np.array(img), np.array([0])])
        except:
            pass
    shuffle(data)
    np.save('data.npy', np.array(data))
```

```python
# fire - [1 0] - 410 images
fire_dir = "Data\\fire"
# nothing - [0 1] - 539 images
no_dir = "Data\\normal" 
img_to_npy(fire_dir, no_dir, img_size = 100)
```
