```python
import subprocess
```

```python
def find_ip(mac_address):
    # Get the list of IP and MAC address in the network
    list_ip = subprocess.check_output(("arp", "-a"))
    list_ip = list_ip.decode("utf-8")
    list_ip = list_ip.splitlines()
    # Find IP of camera
    try:
        for line in list_ip:
            if line.find(mac_address) != -1:
                # extract the ip address
                ip_address = line[2:14]
        return ip_address
    except:
        print("Can't find ip address from the network!")
    
```

```python
# Raspberry Pi 
pi_address = "b8-27-eb-c3-cc-5a"
```

```python
find_ip(pi_address)
```

```python
cam_address = "50-8f-4c-58-71-ff"
```

```python
find_ip(cam_address)
```
