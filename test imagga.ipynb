```python
# Upload image to the imgga API
# Return image_upload_id for classify
def upload(image_path='new.jpg'):
    upload_response = requests.post('https://api.imagga.com/v2/uploads',
                                    auth =(api_key, api_secret),
                                    files={'image':open(image_path, 'rb')})
    data = upload_response.json()
    if data["status"]["type"] == 'success':
        return data["result"]["upload_id"]
    else:
        raise ValueError('Something went wrong')
    # print(json.dumps(upload_response.json(), indent = 2))
    # type(upload_response.json()) output a python dictionary not a json file
```

```python
def classify_image(image_upload_id):
    detected_response = requests.get('https://api.imagga.com/v2/tags?image_upload_id=%s' %image_upload_id,
                auth=(api_key, api_secret))
    print(json.dumps(detected_response.json(), indent = 2))
    return detected_response
```
