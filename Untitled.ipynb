```python
import subprocess
import requests
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow import keras
```

```python
def find_ip():
    # Get the list of IP and MAC address in the network
    list_ip = subprocess.check_output(("arp", "-a"))
    list_ip = list_ip.decode("utf-8")
    list_ip = list_ip.splitlines()
    # My camera MAC address
    mac_add = "50-8f-4c-58-71-ff"
    # Find IP of camera
    for line in list_ip:
        if line.find(mac_add) != -1:
            # extract the ip address
            ip_address = line[2:14]
    return ip_address
```

```python
def take_img(ip_address):
    url = "http://"+ ip_address + ":8080/shot.jpg"
    response = requests.get(url)
    # Decode to get image from byte value
    img = np.array(bytearray(response.content), dtype=np.uint8)
    img = cv2.imdecode(img, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Resize to feed to the network
    img = cv2.resize(img, (50, 50))
    img = np.array([img])
    return img
```

```python
plt.imshow(take_img(find_ip())[0])
plt.title('my picture')
```

```python
model = keras.models.load_model('model\\model_1.h5')
```

```python
model.predict(take_img(find_ip()))
```
